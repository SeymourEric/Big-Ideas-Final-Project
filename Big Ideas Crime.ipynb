```python
import pandas as pd
crime = pd.read_csv("044f2016-1dfd-4ab0-bc1e-065da05fca2e.csv")

crime.head()
```

```python
crimeCount = dict()
i = 0
while i < len(crime):
    neighborhood = crime.loc[i,'INCIDENTNEIGHBORHOOD']
    if neighborhood in crimeCount.keys():
        crimeCount[neighborhood] += 1
    else: 
        crimeCount[neighborhood] = 1
    i += 1

print(crimeCount)
```
